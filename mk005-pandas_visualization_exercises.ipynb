**<p>Recreate this scatter plot of b vs a. Note the color and size of the points. Also note the figure size. See if you can figure out how to stretch it in a similar fashion.</p>**

```python
df3.plot.scatter(x="a", y="b", edgecolor="k", figsize=(10, 3), color="r", s=50)
```

**<p>Create a histogram of the 'a' column like the image below.</p>**
<img src='mk005-pandas_visualization_exercises/pd002.png' /></a>

```python
df3["a"].plot.hist(bins=10, edgecolor="k", color="b")
```

**<p>Create a histogram of the 'a' column like the image below.</p>**
<img src='mk005-pandas_visualization_exercises/pd003.png' /></a>

```python
df3["a"].plot.hist(bins=25, edgecolor="k", color="r", alpha=0.6)
```

**<p>Create a boxplot comparing the a and b columns.</p>**
<img src='mk005-pandas_visualization_exercises/pd004.png' /></a>

```python
df3[["a", "b"]].plot.box(color="r")
```

**<p>Create a kde plot of the 'd' column.</p>**
<img src='mk005-pandas_visualization_exercises/pd005.png' /></a>

```python
df3["d"].plot.density(color="r")
```

**<p>Figure out how to increase the linewidth and make the linestyle dashed.</p>**
<img src='mk005-pandas_visualization_exercises/pd006.png' /></a>

```python
df3["d"].plot.density(color="r", linestyle="--", linewidth=5)
```

**<p>Create an area plot of all the columns for just the rows up to 30.</p>**
<img src='mk005-pandas_visualization_exercises/pd007.png' /></a>
