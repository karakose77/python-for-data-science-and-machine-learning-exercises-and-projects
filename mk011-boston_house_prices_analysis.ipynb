```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(palette="bright")
from sklearn.datasets import load_boston
```

```python
# REAL DATA FROM SKLEARN MODULE
boston = load_boston()
```

```python
boston_df = pd.DataFrame(boston['data'], columns=boston['feature_names'])
boston_df["PRICE"] = boston['target']
boston_df.head()
```

```python
boston_df.info()
```

```python
boston_df.describe()
```

```python
sns.pairplot(boston_df)
```

```python
sns.distplot(boston_df["PRICE"])
```

```python
plt.figure(figsize=(15, 8))
sns.heatmap(boston_df.corr(), annot=True, cmap="coolwarm")
```

```python
boston_Xdf = boston_df[['CRIM', 
                        'ZN', 
                        'INDUS', 
                        'CHAS', 
                        'NOX', 
                        'RM', 
                        'AGE', 
                        'DIS', 
                        'RAD', 
                        'TAX', 
                        'PTRATIO', 
                        'B', 
                        'LSTAT']]
boston_ydf = boston_df['PRICE']
```

```python
from sklearn.model_selection import train_test_split
```

```python
boston_X_train, boston_X_test, boston_y_train, boston_y_test = train_test_split(boston_Xdf, 
                                                                                boston_ydf)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
boston_lm = LinearRegression()
```

```python
boston_lm.fit(boston_X_train, boston_y_train)
```

```python
boston_lm.intercept_
```

```python
boston_lm.coef_
```

```python
boston_lm_df = pd.DataFrame(boston_lm.coef_, 
                            index=boston_X_train.columns, 
                            columns=["Coefficient"])
boston_lm_df
```

```python
boston_predictions = boston_lm.predict(boston_X_test)
```

```python
plt.scatter(boston_y_test, boston_predictions, edgecolors="w")
```

```python
sns.distplot(boston_y_test-boston_predictions)
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(boston_y_test, boston_predictions)
```

```python
metrics.mean_squared_error(boston_y_test, boston_predictions)
```

```python
np.sqrt(metrics.mean_squared_error(boston_y_test, boston_predictions))
```
