# College Data Analysis Project 

For this project we will attempt to use K-Means Clustering to cluster Universities into to two groups, Private and Public. We actually have the labels for this data set, but we will not use them for the K-Means clustering algorithm, since that is an unsupervised learning algorithm. When using the K-Means algorithm under normal circumstances, it is because we don't have labels. In this case we will use the labels to try to get an idea of how well the algorithm performed, but we won't usually do this for K-Means. So the classification report and confusion matrix at the end of this project, don't truly make sense in a real world setting.

## K-Means Cluster Creation

Now it is time to create the Cluster labels.

**Import KMeans from SciKit Learn.**

```python
from sklearn.cluster import KMeans
```

**Create an instance of a K Means model with 2 clusters.**

```python
collage_kmeans = KMeans(n_clusters=2)
```

**Fit the model to all the data except for the Private label.**

```python
collage_kmeans.fit(collage_data.drop(["Private"], axis=1))
```

**What are the cluster center vectors?**

```python
collage_kmeans.cluster_centers_
```

## Evaluation

There is no perfect way to evaluate clustering if you don't have the labels. However since this is just an exercise, we do have the labels. So we take advantage of this to evaluate our clusters.

**Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.**

```python
def converter(private):
    return private == "No"
```

```python
collage_data["Cluster"] = collage_data["Private"].apply(converter)
collage_data.head()
```

**Create a confusion matrix and classification report to see how well the K-Means clustering worked without being given any labels.**

```python
collage_kmeans.labels_
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(collage_data["Cluster"], collage_kmeans.labels_))
print(confusion_matrix(collage_data["Cluster"], collage_kmeans.labels_))
print(accuracy_score(collage_data["Cluster"], collage_kmeans.labels_))
```
