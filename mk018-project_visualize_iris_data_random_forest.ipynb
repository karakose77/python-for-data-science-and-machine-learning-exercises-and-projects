# Visualize a Random Forest

## Import Libraries

```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydot
import random
```

## Load data
Data can also be read from a .csv file.

```python
iris = load_iris()
```

## Create an instance of the random forest model

```python
# Can also use single decision tree
model = RandomForestClassifier(n_estimators=10)
```

## Train the model

```python
model.fit(iris.data, iris.target)
```

## Extract single tree

```python
estimator = model.estimators_[random.randint(1, 9)]
```

## Export as dot file

```python
export_graphviz(estimator, 
                out_file='mk018-project_visualize_iris_data_random_forest/tree.dot', 
                feature_names = iris.feature_names,
                class_names = iris.target_names,
                rounded = True, 
                proportion = False, 
                precision = 2, 
                filled = True)
```

## Convert .dot to .png 

```python
(graph,) = pydot.graph_from_dot_file('mk018-project_visualize_iris_data_random_forest/tree.dot')
graph.write_png('mk018-project_visualize_iris_data_random_forest/tree.png')
```

## Display the .png image

```python
Image(filename = 'mk018-project_visualize_iris_data_random_forest/tree.png')
```
