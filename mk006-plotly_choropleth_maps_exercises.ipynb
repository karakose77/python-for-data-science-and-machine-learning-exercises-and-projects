# Choropleth Maps Exercise 

Welcome to the Choropleth Maps Exercise! In this exercise we will give you some simple datasets and ask you to create Choropleth Maps from them.

```python
import pandas as pd
import plotly.plotly as py
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True)
```

### State_Province_Codes

```python
df1 = pd.read_html('https://www.ups.com/worldshiphelp/WS16/ENU/AppHelp/Codes/State_Province_Codes.htm')
```

```python
t = [df1[0][0][idx] for idx in range(len(df1[0][0])) if "US" in df1[0][0][idx]]
```

```python
l = [df1[0][1][idx] for idx in range(len(df1[0][0])) if "US" in df1[0][0][idx]]
```

```python
zc = list(range(1, 52))
```

```python
data1 = dict(type="choropleth", 
             locations=l, 
             locationmode="USA-states", 
             colorscale="Greens", # "Jet", "Greens" 
             text=t, 
             z=zc, 
             colorbar={"title": "Colorbar"})
```

```python
layout1 = dict(title="US State Codes",
               geo={"scope": "usa"})
```

```python
choromap1 = go.Figure(data=[data1], layout=layout1)
```

```python
iplot(choromap1)
```

```python
plot(choromap1, filename='mk006-plotly_choropleth_maps_exercises_html_outputs/US-state-codes.html')
```

### 2011_US_AGRI_Exports

```python
df2 = pd.read_csv("mk006-plotly_choropleth_maps_exercises/2011_US_AGRI_Exports")
```

```python
df2.head()
```

```python
data2 = dict(type="choropleth", 
             locations=df2["code"], 
             locationmode="USA-states", 
             colorscale="Portland", # "Jet", "Greens" 
             text=df2["text"], 
             z=df2["total exports"],
             marker=dict(line=dict(color="rgb(255, 255, 255)", width=1)),
             colorbar={"title": "Millions USD"})
```

```python
layout2 = dict(title="2011 US Agriculture Exports by State",
               geo=dict(scope="usa",
                        showlakes=True,
                        lakecolor="rgb(85, 173, 240)"))
```

```python
choromap2 = go.Figure(data=[data2], layout=layout2)
```

```python
iplot(choromap2)
```

```python
plot(choromap2, filename='mk006-plotly_choropleth_maps_exercises_html_outputs/US-agriculture-exports.html')
```

### 2014_World_GDP

```python
df3 = pd.read_csv("mk006-plotly_choropleth_maps_exercises/2014_World_GDP")
```

```python
df3.head()
```

```python
data3 = dict(type="choropleth", 
             locations=df3["CODE"],  
             text=df3["COUNTRY"],
             colorscale="Jet",
             z=df3["GDP (BILLIONS)"],
             colorbar={"title": "Billions USD"})
```

```python
layout3 = dict(title="2014 Global GDP",
               geo=dict(showframe=False,
                        projection={"type": "natural earth"}))

# projection type:
# [
# 'equirectangular', 'mercator', 'orthographic', 'natural earth', 'kavrayskiy7', 'miller', 'robinson', 'eckert4', 
# 'azimuthal equal area', 'azimuthal equidistant', 'conic equal area', 'conic conformal', 'conic equidistant', 'gnomonic', 
# 'stereographic', 'mollweide', 'hammer', 'transverse mercator', 'albers usa', 'winkel tripel', 'aitoff', 'sinusoidal'
# ]

# colorscales:
# [
# 'Greys', 'YlGnBu', 'Greens', 'YlOrRd', 'Bluered', 'RdBu', 'Reds', 'Blues', 'Picnic', 'Rainbow', 'Portland', 'Jet', 'Hot', 
# 'Blackbody', 'Earth', 'Electric', 'Viridis', 'Cividis'
# ]
```

```python
choromap3 = go.Figure(data=[data3], layout=layout3)
```

```python
iplot(choromap3)
```

```python
plot(choromap3, filename='mk006-plotly_choropleth_maps_exercises_html_outputs/global-GDP-2014.html')
```

### 2014_World_Power_Consumption

```python
df_pc = pd.read_csv("mk006-plotly_choropleth_maps_exercises/2014_World_Power_Consumption")
df_pc.head()
```

```python
data_pc = dict(type="choropleth", 
               locationmode="country names",
               locations=df_pc["Country"],  
               text=df_pc["Country"], 
               z=df_pc["Power Consumption KWH"],
               colorbar={"title": "KWH"},
               colorscale="Viridis",
               reversescale=True)

layout_pc = dict(title="2014 World Power Consumption",
                 geo=dict(showframe=False, projection={"type": "natural earth"}))
```

```python
choromap4 = go.Figure(data=[data_pc],layout=layout_pc)
iplot(choromap4, validate=False)
```

```python
plot(choromap4, filename='mk006-plotly_choropleth_maps_exercises_html_outputs/world-power-consumption-2014.html')
```

### 2012_Election_Data

```python
df_el = pd.read_csv("mk006-plotly_choropleth_maps_exercises/2012_Election_Data")
df_el.head()
```

**Citizen % is 100% minus non-citizen % in the  "% Non-citizen" column.**

```python
z_el = [str(100 - float(df_el["% Non-citizen"][idx].split("%")[0]))+"% Citizen" for idx in range(len(df_el["% Non-citizen"]))]
```

```python
z_el[:10]
```

```python
data_el = dict(type="choropleth", 
               locations=df_el["State Abv"], 
               locationmode="USA-states", 
               colorscale="Viridis", # "Portland", "Jet", "Greens" 
               reversescale=True,
               text=df_el["State"] + ": " + z_el, 
               z=df_el["Voting-Age Population (VAP)"],
               colorbar={"title": "Voting Age Population"})

layout_el = dict(title="2012 US Voting Age Population by State",
                 geo=dict(scope="usa", showlakes=True, lakecolor="rgb(85, 173, 240)"))
```

```python
choromap5 = go.Figure(data=[data_el], layout=layout_el)
iplot(choromap5, validate=False)
```

```python
plot(choromap5, filename='mk006-plotly_choropleth_maps_exercises_html_outputs/voting-age-population-US-2012.html')
```
